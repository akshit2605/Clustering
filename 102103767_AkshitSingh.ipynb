```python
from pycaret.clustering import *

def modelfunction(clustering_algo='kmeans',transformation = False, normalize = False, pca = False,
      normalize_method = 'zscore',
      transformation_method = 'yeo-johnson',
      pca_method = 'linear', verbose=False):

  setup(data = buddydata, transformation = transformation, normalize = normalize, pca = pca,
      normalize_method = 'zscore',
      transformation_method = 'yeo-johnson',
      pca_method = 'linear', verbose=False)

  print("For Cluster = 3")
  x1 = create_model(clustering_algo, num_clusters = 3)
  metric1=pull()

  print("For Cluster = 4")
  x2 = create_model(clustering_algo, num_clusters = 4)
  metric2=pull()

  print("For Cluster = 5")
  x3 = create_model(clustering_algo, num_clusters = 5)
  metric3=pull()

  newtable=pd.concat([metric1,metric2,metric3],axis=0)
  newtable=newtable.T
  columnname=['Cluster1','Cluster2','Cluster3']
  newtable.columns=columnname
  print(newtable)
  return newtable
```

```python
x=modelfunction()
print(x)
y=modelfunction(clustering_algo='hclust',normalize=True)
print(y)
```

```python
cluster_algo=['kmeans','birch','hclust']

list=[]

for i in range(len(cluster_algo)):

  newtable=pd.DataFrame()

  keys=['With No Data PreProcessing','Normalization','Transformation','PCA','Normalize+Tranforme','Normalize+Transform+PCA']

  # With No Data Preprocessing
  a=modelfunction(clustering_algo=cluster_algo[i])
  # With Normalization
  b=modelfunction(clustering_algo=cluster_algo[i],normalize=True)
  # With Transformation
  c=modelfunction(clustering_algo=cluster_algo[i],transformation = True)
  # wITH pca
  d=modelfunction(clustering_algo=cluster_algo[i],pca = True)
  # With N+T
  e=modelfunction(clustering_algo=cluster_algo[i],normalize=True,transformation = True)
  # with T+N+PCA
  f=modelfunction(clustering_algo=cluster_algo[i],normalize=True,transformation = True,pca=True)

  newtable=pd.concat([a,b,c,d,e,f],axis=1,keys=keys)
  list.insert(i,newtable)
```

```python
Cluster_names=['Kmeans Clustering','Birch Clustering','Hierarchial Clustering']
i=0


with pd.ExcelWriter('result.xlsx') as writer:
  for i, df in enumerate(list):
      print(f"Clustering Algo {i+1}:")
      print(Cluster_names[i])
      print()
      df.to_excel(writer, sheet_name=Cluster_names[i])

print()
print("Result of each Clustering Algorithm on Separate Sheet")
```
